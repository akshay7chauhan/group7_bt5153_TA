```python
import pandas as pd
```

```python
df = pd.read_csv('singapore_trip_advisor_data_sentiment_score.csv')
```

```python
user_profile = df.groupby(['TRIP_TYPE','HOME_COUNTRY']).size().reset_index().rename(columns={0:'count'})
```

```python
user_profile['profile'] = user_profile['HOME_COUNTRY']+'_'+user_profile['TRIP_TYPE']
user_profile
```

```python
df['profile'] = df['HOME_COUNTRY']+'_'+df['TRIP_TYPE']
```

```python
new_df = pd.DataFrame(df.groupby(['profile','ATTRACTION_NAME'])['body_compound'].mean()).reset_index()
new_df
```

```python
new_df2 = pd.merge(left = new_df,right = df,on=['profile','ATTRACTION_NAME'])
new_df2 = new_df2[['profile','ATTRACTION_NAME','body_compound_x']]
new_df2
```

```python
df3 = new_df2.pivot_table(index='ATTRACTION_NAME',columns='profile',values='body_compound_x')
df3
```

```python
df3.fillna(0,inplace=True)
df3.head()
```

```python
from scipy.sparse import csr_matrix
csr_data = csr_matrix(df3.values)
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10, n_jobs=-1)
knn.fit(csr_data)
```

```python
### naive recommendation system
```

```python
def recommender_system_attraction(attraction_name,n_attraction_to_recommend):
    df3.reset_index(inplace=True)
    attraction_name = attraction_name
    n_attraction_to_recommend = n_attraction_to_recommend
    attraction_index = df3[df3['ATTRACTION_NAME'] == attraction_name].index[0]
    distances,indices = knn.kneighbors(csr_data[attraction_index],n_neighbors=n_attraction_to_recommend+1) 
    rec_dict = {
        'Attactions_recommended':list(df3.iloc[indices[0][1:]]['ATTRACTION_NAME'])                          
        }
    return pd.DataFrame(rec_dict)
```

```python
recommender_system_attraction("MacRitchie Nature Trail",3)
```

```python
csr_data_profile = csr_matrix(df3.set_index('ATTRACTION_NAME').T.values)
```

```python
from sklearn.neighbors import NearestNeighbors
knn1 = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10, n_jobs=-1)
knn1.fit(csr_data_profile)
```

```python
### user_profile closness
def recommender_system_user(user_profile,n_close_profiles):
    user_profile = user_profile
    n_close_profiles = n_close_profiles
    filter_df = df3.set_index('ATTRACTION_NAME').T.reset_index()
    profile_index = filter_df[filter_df['profile']==user_profile].index[0]
    distances,indices = knn1.kneighbors(csr_data_profile[profile_index],n_neighbors=n_close_profiles+1) 
    rec_dict = {
        'similar_profiles':list(filter_df.iloc[indices[0][1:]]['profile'])                          
        }
    return pd.DataFrame(rec_dict)
```

```python
recommender_system_user('singapore_Friends',5)
```

```python
recommender_system_attraction("MacRitchie Nature Trail",3)
```
