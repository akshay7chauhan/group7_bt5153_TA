### Wordcloud of Review Body & Review Title Group by Attraction Type

#### Leisure & Recreation

```python
# Review Body
text_leisure = ' '.join(df[df['ATTRACTION_TYPE']=='Leisure & Recreation'].clean_text)
stylecloud.gen_stylecloud(text = text_leisure,
                         icon_name='fas fa-suitcase')
Image(filename='stylecloud.png')
```

```python
# Review Title
text_leisure = ' '.join(df[df['ATTRACTION_TYPE']=='Leisure & Recreation'].REVIEW_TITLE)
stylecloud.gen_stylecloud(text = text_leisure,
                         icon_name='fas fa-plane')
Image(filename='stylecloud.png')
```
